```python
from nba_api.stats.static import players, teams
from nba_api.stats.endpoints import leaguegamefinder
import pandas as pd
```

```python
# ANY TEAMS GAME LOG SINCE 1983

# Get the list of all NBA teams
nba_teams = teams.get_teams()
teams_df = pd.DataFrame(nba_teams)
# Rename the 'id' column to 'teamID'
teams_df.rename(columns={'id': 'TeamID'}, inplace=True)

# can be used to calculate franchise wins/losses etc
# can also break it down by season_id to get gamelog just for this szn
team_abbreviation = "PHI"

# Get the team ID
team_id = teams_df.loc[teams_df['abbreviation'] == team_abbreviation, 'TeamID'].iloc[0]

# Fetch the game log for the team using the obtained team ID
team_game_log = leaguegamefinder.LeagueGameFinder(team_id_nullable=team_id).get_data_frames()[0]
team_game_log
```

```python
# GET GAME LOG FOR JUST THE CURRENT REGULAR SEASON

# Get the list of all NBA teams
nba_teams = teams.get_teams()
teams_df = pd.DataFrame(nba_teams)

# Rename the 'id' column to 'teamID'
teams_df.rename(columns={'id': 'TeamID'}, inplace=True)

# Specify the team abbreviation
team_abbreviation = "PHI"

# Get the team ID
team_id = teams_df.loc[teams_df['abbreviation'] == team_abbreviation, 'TeamID'].iloc[0]

# Fetch the game log for the team using the obtained team ID
team_game_log = leaguegamefinder.LeagueGameFinder(team_id_nullable=team_id).get_data_frames()[0]

# Filter the game log for the regular season (games after October 26, 2023)
regular_season_games = team_game_log[team_game_log['GAME_DATE'] >= '2023-10-24']

# Print the game log for the regular season
regular_season_games
```
